# Inital Cleaning
This notebook performs the initial preparation of the raw dataset to ensure it is ready for further analysis and modeling.  
The main steps include:

- Converting data types (e.g., dates, times, numeric fields)
- Translating column names and values from Hebrew to English
The output of this stage is a consistent and clean dataset that will be used for feature engineering.

### Imports

```python
import pandas as pd
import sys
from pathlib import Path
sys.path.append(str(Path("..").resolve()))
from src.config import COLS_HEB_TO_ENG,HEB_VALS_TO_ENG
from src.data_preparation import rename_columns, convert_data_types, translate_columns
```

### Process

```python
### Import raw data
df = pd.read_csv(r'../data/raw_data/sample/telaviv_buses_jan2024_sample_route.csv')

### Rename the columns names to english
df = rename_columns(df,COLS_HEB_TO_ENG)

### Convert and ensure data types
convert_data_types(df)

### Translate hebrew values to english
df = translate_columns(df, ['day','agency_name','route_type'], HEB_VALS_TO_ENG)
```

### Export cleaned DF

```python
df.to_csv(r'../data/interim_data/telaviv_buses_jan2024_sample_route_cleaned.csv',index=False,encoding='utf-8-sig')
```
