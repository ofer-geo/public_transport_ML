```python
def build_probability_table(
    df,
    bins=[0, 30000, 100000,float('inf')],
    labels=['0-30k', '30-100k','100k+'],
    smoothing_alpha=0
):
    import numpy as np
    
    df = df.copy()

    # Create bins
    df['route_length_bin'] = pd.cut(
        df['route_length'],
        bins=bins,
        labels=labels,
        right=False
    )

    # Count occurrences
    counts = (
        df.groupby(['hour_rounded', 'route_length_bin', 'target'], observed=True)
          .size()
          .unstack(fill_value=0)
    )

    # --- Smoothing (optional but recommended) ---
    if smoothing_alpha > 0:
        global_dist = df['target'].value_counts(normalize=True)
        counts = counts + smoothing_alpha * global_dist

    # Convert to probabilities
    prob_table = counts.div(counts.sum(axis=1), axis=0)

    return prob_table
```

```python
prob_table = build_probability_table(df, smoothing_alpha=0)
```

```python
def apply_early_probability(
    df,
    prob_table,
    bins=[0, 30000, 100000, float('inf')],
    labels=['0-30k', '30-100k','100k+']
):
    df = df.copy()

    # Create same bins (IMPORTANT: must match training)
    df['route_length_bin'] = pd.cut(
        df['route_length'],
        bins=bins,
        labels=labels,
        right=False
    )

    # Extract only early probability
    early_prob = prob_table['early'].reset_index()

    # Merge
    df = df.merge(
        early_prob,
        on=['hour_rounded', 'route_length_bin'],
        how='left'
    )

    # Rename column
    df = df.rename(columns={'early': 'early_by_hour_length_proba'})

    # Handle missing values (important!)
    global_mean = early_prob['early'].mean()
    df['early_by_hour_length_proba'] = df['early_by_hour_length_proba'].fillna(global_mean)

    return df
```

```python
import numpy as np
import pandas as pd

bins = [0, 30000, 50000, 70000, np.inf]
labels = ['0-30k', '30-50k', '50-70k', '70k+']

def add_route_length_bin(df):
    df = df.copy()
    df['route_length_bin'] = pd.cut(
        df['route_length'],
        bins=bins,
        labels=labels,
        right=False
    )
    return df


def build_hybrid_early_probability_tables(
    df_train,
    min_count_strong=50,
    alpha=5
):
    df_train = add_route_length_bin(df_train)

    global_early = (df_train['target'] == 'early').mean()

    # 1. Detailed: hour + length + agency
    detailed_counts = (
        df_train
        .groupby(['hour_rounded', 'route_length_bin', 'agency_name'], observed=True)
        .agg(
            total=('target', 'size'),
            early=('target', lambda x: (x == 'early').sum())
        )
        .reset_index()
    )

    detailed_counts['prob_detailed'] = (
        (detailed_counts['early'] + alpha * global_early) /
        (detailed_counts['total'] + alpha)
    )

    detailed_counts.loc[
        detailed_counts['total'] < min_count_strong,
        'prob_detailed'
    ] = np.nan

    # 2. Fallback: hour + length
    hour_len_counts = (
        df_train
        .groupby(['hour_rounded', 'route_length_bin'], observed=True)
        .agg(
            total=('target', 'size'),
            early=('target', lambda x: (x == 'early').sum())
        )
        .reset_index()
    )

    hour_len_counts['prob_hour_len'] = (
        (hour_len_counts['early'] + alpha * global_early) /
        (hour_len_counts['total'] + alpha)
    )

    # 3. Fallback: agency only
    agency_counts = (
        df_train
        .groupby('agency_name')
        .agg(
            total=('target', 'size'),
            early=('target', lambda x: (x == 'early').sum())
        )
        .reset_index()
    )

    agency_counts['prob_agency'] = (
        (agency_counts['early'] + alpha * global_early) /
        (agency_counts['total'] + alpha)
    )

    return {
        'detailed': detailed_counts,
        'hour_len': hour_len_counts,
        'agency': agency_counts,
        'global_early': global_early
    }


def apply_hybrid_early_probability(df, tables):
    df = add_route_length_bin(df)

    df = df.merge(
        tables['detailed'][[
            'hour_rounded', 'route_length_bin', 'agency_name', 'prob_detailed'
        ]],
        on=['hour_rounded', 'route_length_bin', 'agency_name'],
        how='left'
    )

    df = df.merge(
        tables['hour_len'][[
            'hour_rounded', 'route_length_bin', 'prob_hour_len'
        ]],
        on=['hour_rounded', 'route_length_bin'],
        how='left'
    )

    df = df.merge(
        tables['agency'][[
            'agency_name', 'prob_agency'
        ]],
        on='agency_name',
        how='left'
    )

    df['early_by_hour_length_agency_proba'] = (
        df['prob_detailed']
        .fillna(df['prob_hour_len'])
        .fillna(df['prob_agency'])
        .fillna(tables['global_early'])
    )

    df = df.drop(columns=['prob_detailed', 'prob_hour_len', 'prob_agency'])

    return df
```

```python
df_copy = df.copy()
```

```python
tables = build_hybrid_early_probability_tables(
    df_copy,
    min_count_strong=50,
    alpha=5
)
```

```python
df_copy = apply_hybrid_early_probability(df_copy, tables)
```
